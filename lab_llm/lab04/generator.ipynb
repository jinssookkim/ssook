# 반복문 안에서의 return

```python
def return_in_for():
    for n in range(10):
        return n
```

`return` 의 의미
*   값을 함수를 호출한 곳으로 반환
*   함수를 종료

```python
result = return_in_for()
print(result)
```

# 반복문 안에서의 yield


```python
def yield_in_for():
    for n in range(10):
        yield n
```

```python
result = yield_in_for()
print(result)
```

*   `yield`를 사용하는 함수의 리턴값은 `generator`라고 하는 객체 타입
*   `generator`객체는 for-in 구문에서 사용하는 iterable 객체.

```python
for x in result:
    print(x)
```

```python
for x in result:
    print(x)
```

*   `generator` 객체는 for-in 구문을 사용해서 yield되는 값들을 순서대로 출력
*   yield되는 값들은 한 번 출력되면 메모리에서 삭제됨
*   첫 번째 for-in 구문에서는 yield되는 값들을 순서대로 출력
*   두 번째 for-in 구문에서는 yield된 값들이 한 번씩 사용이 완료되어 출력할 값이 존재하지 않음

```python
# 얘는 얼마든지 사용 가능
for x in yield_in_for():
    print(x)
```

```python
def my_generator():
    values=['Google',
            'Meta',
            'OpenAI',
            'Anthropic']

    for v in values:
        yield v
```

```python
for x in my_generator():
    print(x)
```

```python
response = client.chat.completions.create(
    model='gpt-5.4-mini',
    temperature=0.9,
    messages=[
        {
            'role': 'user',
            'content': '2026년 이란 전쟁에 대해서 요약해줘.'
        }
    ],
    stream=True
)

for chunk in response:
    print(chunk.choices[0].delta.content, end='')
    #> 가장 마지막 chunk(파편화된 답변 조각)의 content는 None.
```
